# Programming in Data Science - TP2 : Building of indicators with Pandas & Python


## 1. Loading the data


```python
# Library imports
import pandas as pd

# display oprions
pd.set_option('display.max_columns', None)
```

```python
# Recovering data
films = pd.read_csv('films.csv', sep=';', skiprows=[1], encoding='ISO-8859-1')
df_films = pd.DataFrame(films)
df_films.head()
```

```python
# Loading of cars df
cars = pd.read_csv('cars.csv', sep=';', skiprows=[1], encoding='ISO-8859-1')
df_cars = pd.DataFrame(cars)
df_cars.head()
```

## 2. Data preprocessing

### 2.1. Films

```python
df_films.describe(include='all')
```

```python
df_films.dtypes
```

```python
df_films['ImageFile'] = df_films['*Image'].astype(str).str.replace('*', '', regex=False)
df_films = df_films.drop(columns=['*Image'])
colonnes_str = ['Title', 'Subject', 'Actor', 'Actress', 'Director', 'Awards', 'ImageFile']

for col in colonnes_str:
    df_films[col] = df_films[col].astype(str).str.strip()
df_films = df_films.dropna(subset=['Year', 'Length'])

df_films.head()
```

```python
print(f"Duplicate : {df_films.duplicated().sum()}")
print(df_films.isnull().sum())
```

### 2.2. Cars

```python
df_cars.describe(include='all')
```

```python
df_cars.dtypes
```

```python
print(f"Duplicate : {df_cars.duplicated().sum()}")
print(df_cars.isnull().sum())
df_cars = df_cars.dropna(subset=['MPG', 'Horsepower'])
df_cars.head()
```

## 3. Key Indicators for the cars data set

### 3.1 Mean power by number of cylinder 

```python
hp_par_cylindre = df_cars.groupby('Cylinders')['Horsepower'].mean().sort_index()

print(hp_par_cylindre)
```

### 3.2 Average consumption by origine

```python
mpg_par_origine = df_cars.groupby('Origin')['MPG'].mean().sort_values(ascending=False)

print(mpg_par_origine)
```

### 3.3 Top 5 heaviest cars

```python
df_cars.sort_values(by='Weight', ascending=False).head(5)
```

## 4. Indicator for the films Data Set

### 4.1 Number of nominated films by subject

```python
films_primes = df_films[df_films['Awards'].str.lower() == 'yes']
films_primes['Subject'].value_counts()
```

### 4.2 Best Director (by popularity)

```python
df_films.groupby('Director')['Popularity'].mean().sort_values(ascending=False).head(10)
```

### 4.3 Number of films by decades

```python
df_films['Decade'] = (df_films['Year'] // 10) * 10
df_films['Decade'].value_counts().sort_index()
```

## 5. Indicator for *Northwind* DataFrame

```python
orders = pd.read_csv('NorthwindDB/orders.csv', on_bad_lines='skip')
customers = pd.read_csv('NorthwindDB/customers.csv', on_bad_lines='skip')
order_details = pd.read_csv('NorthwindDB/order-details.csv')
products = pd.read_csv('NorthwindDB/products.csv')
categories = pd.read_csv('NorthwindDB/categories.csv')
employees = pd.read_csv('NorthwindDB/employees.csv')
```

### 5.1 Number of orders by country

```python
cmd_cli = orders.merge(customers, on='customerID')
cmd_cli.groupby('country')['orderID'].nunique().sort_values(ascending=False).head(10)
```

### 5.2 Local sales sorted by categories

```python
od_prod = order_details.merge(products, on='productID', suffixes=('_order', '_product'))
od_prod_cat = od_prod.merge(categories, on='categoryID')

od_prod_cat['ventes'] = od_prod_cat['quantity'] * od_prod_cat['unitPrice_order']

sales_by_cat = (
    od_prod_cat.groupby('categoryName')['ventes']
    .sum()
    .sort_values(ascending=False)
)

print("Total Sales by Categories :")
print(sales_by_cat)
```

### 5.3 Top 5 employees 

```python
cmd_det = orders.merge(order_details, on='orderID')
cmd_det['ventes'] = cmd_det['quantity'] * cmd_det['unitPrice']
ventes_emp = cmd_det.merge(employees, on='employeeID')
ventes_emp.groupby(['employeeID','lastName','firstName'])['ventes'].sum().sort_values(ascending=False).head(5)
```
